### Here Data is heavily imbalanced so we need to introduce randomness in the datasets since the possibillity of faud being detected as not fraud is way higher. Which will decrease the model performance.

```python
# Fetching the output columns fraud and not_fraud to plot the graph
fraud = df[df['is_fraud']==1]
not_fraud = df[df['is_fraud']==0]
```

```python
print(fraud.shape[0])
print(not_fraud.shape[0])
```

```python
not_fraud = not_fraud.sample(fraud.shape[0], random_state=40)
df_new = pd.concat([not_fraud, fraud], axis=0)
```
